```python
import numpy as np
import pandas as pd
import xgboost as xgb
from xgboost import plot_importance, plot_tree
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split, GridSearchCV
from gensim.models import Word2Vec
import spacy
nlp = spacy.load('en_core_web_sm')
```

```python
df = pd.read_csv('clean_data.csv')
```

```python
df.drop(columns=df.columns[0], axis=1, inplace=True)
df
```

```python
sentences = [str(i).split() for i in df['processed_tweet']]
model = Word2Vec(sentences, vector_size=300,window=10, min_count=2, sg  = 0)
```

```python
def sentence_emb(sentence, model):
    vectors = [model.wv[word] for word in sentence if word in model.wv]
    if len(vectors)==0:
        return None
    sentence_e = sum(vectors)/len(vectors)
    return sentence_e
```

```python
sentence_embeddings = [sentence_emb(sentence,model) for sentence in sentences]
```

```python
df['embedded'] = sentence_embeddings
```

```python
df.dropna(inplace=True)
```

```python
leng = len(list(df['embedded']))
x_train = list(df['embedded'])[0 : int(0.7 * leng) ]
x_val = list(df['embedded'])[int(0.7 * leng) : int(0.8* leng)]
y_val = list(df['class'])[int(0.7 * leng) : int(0.8 * leng)]
x_test = list(df['embedded'])[int(0.8*leng):]
y_train = list(df['class'])[0 : int(0.7 * leng) ]
y_test = list(df['class'])[int(0.8*leng):]
```

```python

xgb_classifier = xgb.XGBClassifier(
    max_depth=5,
    learning_rate=0.1,
    n_estimators=100
)

xgb_classifier.fit(x_train, y_train)
```

```python
text = "I hate you. I don't like you"
def preprocess_token(token):
    return token.lemma_.strip().lower()
def valid_token(token):
    return bool(token and token.lemma_.isalnum() and str(token).strip() and not token.is_punct)

doc = nlp(text)

processed = [preprocess_token(token) for token in doc if valid_token(token)]
```

```python
text_emb = sentence_emb(processed, model)
```

```python
parameters = {
    'n_estimators': [100,200,300,400],
    'learning_rate': [0.001, 0.005, 0.01, 0.05],
    'max_depth': [8,10,12,15],
    'gamma': [0.001, 0.005, 0.01, 0.02],
    'random_state' : [42]
}
eval_set = [(x_train,y_train), (x_val,y_val)]
Modell = xgb.XGBClassifier(eval_set=eval_set)
clf = GridSearchCV(Modell,parameters)
clf.fit(x_train,y_train)

print(f'Best params: {clf.best_params_}')
print(f'Best validation score = {clf.best_score_}')
```
